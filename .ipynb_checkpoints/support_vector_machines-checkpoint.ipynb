# Support Vector Machines

## Support Vector Classifier

```python
from sklearn.svm import SVC as SKLSVC
from ml.svm import SVC
```

### Linear Kernel

```python
from ml.kernels import linear_kernel
```

#### Hard Margin

```python
from utils import generate_linearly_separable_overlap_data
from sklearn.model_selection import train_test_split

X, y = generate_linearly_separable_overlap_data()
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75)

svc = SVC(kernel=linear_kernel).fit(X_train, y_train)
print('custom svc accuracy score: ' + str(svc.score(X_test, y_test)))
SVC.plot(svc, X_train, y_train)

svc = SKLSVC(kernel='linear').fit(X_train, y_train)
print('sklearn svc accuracy score: ' + str(svc.score(X_test, y_test)))
SVC.plot(svc, X_train, y_train)
```

#### Soft Margin

```python
from utils import generate_linearly_separable_data

X, y = generate_linearly_separable_data()
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75)

svc = SVC(kernel=linear_kernel, C=0.1).fit(X_train, y_train)
print('custom svc accuracy score: ' + str(svc.score(X_test, y_test)))
SVC.plot(svc, X_train, y_train)

svc = SKLSVC(kernel='linear', C=0.1).fit(X_train, y_train)
print('sklearn svc accuracy score: ' + str(svc.score(X_test, y_test)))
SVC.plot(svc, X_train, y_train)
```

### Polynomial Kernel

```python
from utils import generate_non_linearly_separable_data
from ml.kernels import polynomial_kernel

X, y = generate_non_linearly_separable_data()
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75)

svc = SVC(kernel=polynomial_kernel).fit(X_train, y_train)
print('custom svc accuracy score: ' + str(svc.score(X_test, y_test)))
SVC.plot(svc, X_train, y_train)

svc = SKLSVC(kernel='poly').fit(X_train, y_train)
print('sklearn svc accuracy score: ' + str(svc.score(X_test, y_test)))
SVC.plot(svc, X_train, y_train)
```

### RBF Kernel

```python
from utils import generate_non_linearly_separable_data
from ml.kernels import rbf_kernel

X, y = generate_non_linearly_separable_data()
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75)

svc = SVC(kernel=rbf_kernel).fit(X_train, y_train)
print('custom svc accuracy score: ' + str(svc.score(X_test, y_test)))
SVC.plot(svc, X_train, y_train)

svc = SKLSVC(kernel='rbf').fit(X_train, y_train)
print('sklearn svc accuracy score: ' + str(svc.score(X_test, y_test)))
SVC.plot(svc, X_train, y_train)
```
